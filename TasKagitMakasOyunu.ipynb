```python
import random
# random kütüphanesinden yararlanacağım için bu kütüphaneyi import ettim
# İlk tanımlamaları yaptım
hamleler = ["taş", "kağıt", "makas"]
oyuncu_puan = 0
robot_puan = 0
```

```python
kac_hamle = input("Kaç hamle oynamak istersin: ")
hamle_sayaci = 0
while True:
  ayrac = "- " * 20
  hamle_sayaci += 1
  print(f"{ayrac}  \n{hamle_sayaci}. Round: ")

  robot_secim = random.choice(hamleler)
  kontrol = True 
  while kontrol:  
    secimin = input("Senin seçimin hangisi? ('taş', 'kağıt', 'makas' sadece birini yaz): ")
    if secimin not in hamleler:
      print("('taş', 'kağıt', 'makas' sadece birini yaz):")
    else:
      kontrol = False

  print(f"Senin seçimin: {secimin} ve robotun seçimi: {robot_secim} olduğu için...")

  if robot_secim == secimin:
    print("Aynı seçimi yaptınız.")

  elif robot_secim == "kağıt":
    if secimin == "taş":
      print("Bu roundu 'robot' kazandı!")
      robot_puan += 1
    else:
      print( "Bu roundu 'sen' kazandın!")
      oyuncu_puan += 1
  elif robot_secim == "taş":
    if secimin == "makas":
      print("Bu roundu 'robot' kazandı!")
      robot_puan += 1
    else:
      print( "Bu roundu 'sen' kazandın!")
      oyuncu_puan += 1
  elif robot_secim == "makas":
    if secimin == "kağıt":
      print("Bu roundu 'robot' kazandı!")
      robot_puan += 1
    else:
      print( "Bu roundu 'sen' kazandın!")
      oyuncu_puan += 1

  if hamle_sayaci == int(kac_hamle):
    print("Oyun Bitti! \nOYUN SONUCU: ")
    break

if robot_puan == oyuncu_puan:
  print(f"Kazanan Yok! Robot ve Sen {oyuncu_puan} puan alarak oyunu berebere bitirdiniz. ")
elif robot_puan > oyuncu_puan:
  print(f"Oyunu kazanan: ROBOT {robot_puan} puan aldı. (senin puanın: {oyuncu_puan})")
else:
  print(f"Oyunu SEN Kazandın: Senin puanın: {oyuncu_puan} (Robot: {robot_puan} puan alabildi)")
```
